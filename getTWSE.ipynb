```python
import pandas as pd
import json
import requests
import os
```

```python
url = 'https://www.twse.com.tw/rwd/zh/afterTrading/BFT41U?date=20240812&selectType=ALLBUT0999&response=json&_=1723714334350'
data = requests.get(url).text
data = json.loads(data)
```

```python
ohlcv_data = pd.DataFrame(data['data'])
ohlcv_data = ohlcv_data[: -1]
time_series = pd.DataFrame([data['date']] * len(ohlcv_data))
fetch_df = pd.concat([ohlcv_data, time_series], axis = 1)
fetch_df.columns = ['symbal', 'name', 'volume', 'count', 'amount', 'close', 'last_display_buy_volume', 'last_display_sell_volume', 'datetime']
fetch_df
```
