```python
failed_dates = [
    "2021-02-02", "2021-03-17", "2021-05-11", "2021-07-14", "2021-07-20",
    "2022-03-16", "2022-12-23",
    "2023-07-19", "2023-08-02", "2023-08-30", "2023-10-11", "2023-10-20",
    "2024-03-12", "2024-04-09", "2024-05-22", "2024-05-24", "2024-06-04",
    "2024-06-05", "2024-10-15",
    "2025-01-06",
]

base = Path("data/TaiwanOptionTick/TXO")

records = []

for d in failed_dates:
    year = d[:4]
    path = base / year / f"{d}.parquet"

    if path.exists():
        print(f"skip existing: {d}")
        continue

    try:
        df = fm.get_option_tick(option_id="TXO", trade_date=d)

        if df.empty:
            print(f"empty: {d}")
            records.append({"date": d, "status": "empty", "rows": 0})
            continue

        path.parent.mkdir(parents=True, exist_ok=True)
        df.to_parquet(path, index=False)

        print(f"saved: {d}, rows={len(df)}")
        records.append({"date": d, "status": "saved", "rows": len(df)})

    except Exception as e:
        print(f"failed: {d}, {e}")
        records.append({"date": d, "status": "failed", "rows": None, "error": str(e)})

retry_manifest = pd.DataFrame(records)
retry_manifest
```
