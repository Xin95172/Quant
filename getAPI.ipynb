# 抓多幣種資料

```python
symbal_list = ['BTCUSDT', 'ETHUSDT', 'BNBUSDT', 'AAVEUSDT', 'ETCUSDT', 'SOLUSDT']
```

```python
def fetch_ohlcv_data(symbal_list: list, interval: str, since, end):
    ohlcv_dict = {}

    for target in symbal_list:
        pair = target
        ContractType = 'PERPETUAL'
        start_time = int(pd.Timestamp(since).timestamp() * 1000)
        end_time = int(pd.Timestamp(end).timestamp() * 1000)
        limit = 1000

        symbal_data = []

        while start_time < end_time:
            print(pd.to_datetime(start_time, unit = 'ms'))
            params = {
                'pair': pair,
                'ContractType': ContractType,
                'interval': interval,
                'startTime': start_time,
                'limit': limit
            }

            url = 'https://fapi.binance.com/fapi/v1/continuousKlines'
            res = requests.get(url, params = params)
            data = res.json()

            if not data:
                break

            symbal_data.extend(data)
            start_time = int(data[-1][0]) + 1
            time.sleep(0.1)

        if symbal_data:
            df = pd.DataFrame({
                'open_datetime': [row[0] for row in data],
                'open': pd.to_numeric([row[1] for row in data]),
                'high': pd.to_numeric([row[2] for row in data]),
                'low': pd.to_numeric([row[3] for row in data]),
                'close': pd.to_numeric([row[4] for row in data]),
                'volume': pd.to_numeric([row[5] for row in data]),
                'close_datetime': pd.to_numeric([row[6] for row in data]),
                'volvalue': pd.to_numeric([row[7] for row in data]),
                'takerbuy': pd.to_numeric([row[8] for row in data]),
                'takerbuyvalue': pd.to_numeric([row[9] for row in data])
            })

            df['open_datetime'] = pd.to_datetime(df['open_datetime'], unit = 'ms')
            df['open_datetime'] = df['open_datetime'].dt.strftime('%Y-%m-%d %H:%M:%S')
            df['close_datetime'] = pd.to_datetime(df['close_datetime'], unit = 'ms')
            df['close_datetime'] = df['close_datetime'].dt.strftime('%Y-%m-%d %H:%M:%S')

            ohlcv_dict[target] = df

    return ohlcv_dict
```

```python
factor_dict = fetch_ohlcv_data(symbal_list = symbal_list,
                               interval = '1d',
                               since = '2024-1-1',
                               end = '2024-8-1')
```

# Get Top Long short position

```python
def fetch_top_ls_position_data(symbal_list, interval):
    top_ls_position_dict = {}

    for target in symbal_list:
        url = 'https://fapi.binance.com/futures/data/topLongShortPositionRatio'
        params = {'symbal': target, 'period': interval, 'limit': '500'}
        res = requests.get(url, params = params)
        res.raise_for_status()
        
        try:
            data = res.json()
        except:
            print('Error: Response is not in JSON format.')
            print('Response content: %s', res.text)
            continue

        target_df = pd.DataFrame(data.copy()).drop(columns = 'symbal')
        target_df['timestamp'] = pd.to_datetime(target_df['timestamp'], unit = 'ms')
        target_df['datetime'] = target_df['timestamp'].dt.strftime('%Y-%m-%d %H:%M:%S')
        target_df = target_df.set_index('datetime').drop(columns = 'timestamp')
        target_df['ToplongAccount'] = pd.to_numeric(target_df['longAccount'])
        target_df['ToplongShortRatio'] = pd.to_numeric(target_df['longShortRatio'])
        target_df['TopshortAccount'] = pd.to_numeric(target_df['shortAccount'])

        top_ls_position_dict[target] = target_df

    return top_ls_position_dict
```
