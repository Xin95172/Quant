```python
import numpy as np
import pandas as pd
import yfinance as yfin
import os
from datetime import timedelta
import requests
from bs4 import BeautifulSoup
import re
from datetime import datetime, timedelta
import time
import random
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import InvalidSessionIdException
```

```python
symbols = []

file_path = '/Users/xinc./Documents/GitHub/Quant/20200102（股票代號）——Heather.xlsx'
df = pd.read_excel(file_path)

for index, row in df.iterrows():
    symbols.append((row['代號'], row['名稱']))
```

```python
updated_symbols = []
failed_symbols = []

driver = webdriver.Chrome()

def fetch_stock_info(code, name):
    
    driver.get(f'https://www.twse.com.tw/zh/listed/profile/company.html?{code}')
    
    try:
        time.sleep(5)
        
        iframe = WebDriverWait(driver, 15).until(
            EC.presence_of_element_located((By.TAG_NAME, "iframe"))
        )
        driver.switch_to.frame(iframe)
    
        industry_name_element = driver.find_element(By.XPATH, "//span[contains(text(), '產業類別')]/preceding::span[1]")
        industry_name = industry_name_element.text.strip()

        updated_symbols.append((code, name, industry_name))
        print(f"成功獲取: {code} - {name} ({industry_name})")

    except Exception as e:
        print(f"{code} 產業類別擷取失敗: {e}")
        failed_symbols.append((code, name))

    driver.switch_to.default_content()
    time.sleep(25)

for code, name in symbols:
    fetch_stock_info(code, name)

while failed_symbols:
    print("開始重試失敗的股票...")
    retry_failed = failed_symbols.copy()
    failed_symbols.clear()

    for code, name in retry_failed:
        fetch_stock_info(code, name)

driver.quit()
```

```python
updated_symbols = []
failed_symbols = []
driver = webdriver.Chrome()

def fetch_stock_info(code, name):
    global driver
    
    try:
        driver.get(f'https://www.twse.com.tw/zh/listed/profile/company.html?{code}')
        time.sleep(5)
        
        iframe = WebDriverWait(driver, 15).until(
            EC.presence_of_element_located((By.TAG_NAME, "iframe"))
        )
        driver.switch_to.frame(iframe)
    
        industry_name_element = driver.find_element(By.XPATH, "//span[contains(text(), '產業類別')]/preceding::span[1]")
        industry_name = industry_name_element.text.strip()

        updated_symbols.append((code, name, industry_name))
        print(f"成功獲取: {code} - {name} ({industry_name})")

    except InvalidSessionIdException:
        print("Session 已失效，重新啟動 driver...")
        driver.quit()
        driver = webdriver.Chrome()
        fetch_stock_info(code, name)

    except Exception as e:
        print(f"{code} 產業類別擷取失敗: {e}")
        failed_symbols.append((code, name))

    driver.switch_to.default_content()
    time.sleep(random.uniform(20, 25))

for code, name in symbols:
    fetch_stock_info(code, name)

while failed_symbols:
    retry_failed = failed_symbols.copy()
    failed_symbols.clear()
    
    for code, name in retry_failed:
        fetch_stock_info(code, name)

driver.quit()
```
