```python
import pandas as pd
import numpy as np
import chardet
```

```python
data = pd.read_csv('data/Google 雲端硬碟數據.txt', encoding = 'Big5', encoding_errors = 'ignore', delimiter='\t')
data
```

```python
data = data.drop(['開盤價', '收盤價'], axis = 1)
data.columns = ['Contract', 'Name', 'Date', 'Settlement', 'Volume']
data.replace('-', np.nan, inplace = True)

numeric_cols = ['Settlement', 'Volume']
for col in numeric_cols:
    data[col] = pd.to_numeric(data[col], errors = 'coerce')

data['Date'] = pd.to_datetime(data['Date'], format = '%Y%m%d')

# data.dropna(inplace = True)
```

```python
data.to_csv('data.csv', index = False)
```
